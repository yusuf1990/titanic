```python
df['Cabin']=df['Cabin'].fillna('N/-1/u')
#df['Cabin'].str.split('/')[0]
df['deck']=df['Cabin'].apply(lambda x:x.split('/')[0])
df['num']=df['Cabin'].apply(lambda x:x.split('/')[1])
df['side']=df['Cabin'].apply(lambda x:x.split('/')[2])
df.head()
```

```python
Deck=pd.get_dummies(df['deck'],drop_first=True)
Side=pd.get_dummies(df['side'],drop_first=True)
```

```python
train=pd.concat([df,Deck,Side],axis=1)
```

```python
train.drop(['deck','side'],axis=1,inplace=True)
```

```python
train['VIP'].fillna('False',inplace=True)
train['Destination'].fillna('Unknown',inplace=True)
train['HomePlanet'].fillna('Unknown',inplace=True)
train['CryoSleep'].fillna(False,inplace=True)
train['Destination'].fillna('Unknown',inplace=True)
train['CryoSleep'].fillna(False,inplace=True)
train['HomePlanet'].fillna('Unknown',inplace=True)
```

```python
homeplanet=pd.get_dummies(train['HomePlanet'],drop_first=True)
cryosleep=pd.get_dummies(train['CryoSleep'],drop_first=True)
cryosleep=cryosleep.rename(columns={True:'CryoTR'})
destination=pd.get_dummies(train['Destination'],drop_first=True)
#transported=pd.get_dummies(train['Transported'],drop_first=True)
#transported=transported.rename(columns={True:'TranTR'})
vip=pd.get_dummies(train['VIP'],drop_first=True)
vip=vip.rename(columns={True:'VIPTR'})
```

```python
train=pd.concat([train,homeplanet,cryosleep,destination,vip],axis=1)
train.drop(['Cabin','HomePlanet','CryoSleep','Destination','VIP'],axis=1,inplace=True)
```
